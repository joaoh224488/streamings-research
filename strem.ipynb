```python
def in_strem(data, str, str2 =None, str3 = None):
    if str2 == None and str3 == None:
        return data.loc[data[str] == 1]
    elif str2 != None and str3 == None:
        return data.loc[(data[str] == 1) | (data[str2] == 1)]
    elif str2 != None and str3 != None:
        return data.loc[(data[str] == 1) | (data[str2] == 1) | (data[str3] == 1)]
```
